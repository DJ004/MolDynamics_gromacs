```python
# plotting XVG file with multiple Y-columns, where 1st column is X data

import matplotlib.pyplot as plt
import numpy as np

xvg_file = 'data.xvg'
y_col = [2, 3]  # 1-based index. Example: plot columns 2,3 as Y data
y_labels = {
    2: 'y_col_2',
    3: 'y_col_3',
}

data = np.loadtxt(xvg_file, comments=['@', '#'])
x_data = data[:, 0]

fig, ax = plt.subplots(1,1, figsize=(7, 4), layout='tight')
cmap_name = 'rainbow'  # hsv, brg, viridis, plasma | matplotlib.org/stable/users/explain/colors/colormaps.html
cmap = plt.cm.get_cmap(cmap_name)
colors = cmap(np.linspace(0, 1, len(y_col)))

for i, col in enumerate(y_col):
    y_data = data[:, col - 1]  # convert to 0-based index
    legend_label = y_labels.get(col, f'Column {col}')
    ax.plot(x_data, y_data, label=legend_label, color=colors[i], marker='o', markersize=3, linestyle='-')
    
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title(xvg_file)
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
ax.grid(alpha=0.3)
plt.show()

# plt.savefig(xvg_file.replace('.xvg', '.svg'), format='svg', dpi=300)
```
